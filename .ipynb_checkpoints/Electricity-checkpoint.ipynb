## Import all the data from 2019 and merge in a DF

```python
list_csv= []
df_elec_2019 = pd.read_csv("data/Electricity/stedin_electricity_2019.csv")
import os
for dirname, _, filenames in os.walk('data/Electricity/', topdown=False):
    for filename in filenames[1:]:
        if '2019.csv' in filename:
            list_csv.append(os.path.join(dirname, filename))
            df = pd.read_csv(os.path.join(dirname, filename))
            df_elec_2019 = df_elec_2019.append(df, ignore_index=True, sort=False)
```

## Create an addresses column from street names, zip codes and city name 

```python
df_elec_2019['address'] = df_elec_2019['street'] + ' ' + df_elec_2019['zipcode_from'] + ' ' + df_elec_2019['city']
```

## Use google API to retrieve latitude and longitude 

```python
import googlemaps
API_key = 'XXXX'
gmaps_keys = googlemaps.Client (key = API_key)
```

```python
#Longitude stored in 'longitude column'
d = {}
for i in range(len(df_elec_2019['address'])): 
    geocoderesult = gmaps_keys.geocode(df_elec_2019['address'][i])
    d[i] =  geocoderesult[0]['geometry']['location']['lng']
    
Lon = pd.DataFrame.from_dict(d, orient='index', columns= ['Lon'])
df_elec_2019['longitude'] = Lon
```

```python
#Latitude stored in 'latitude column'
d = {}
for i in range(len(df_elec_2019['address'])): 
    geocoderesult = gmaps_keys.geocode(df_elec_2019['address'][i])
    d[i] =  geocoderesult[0]['geometry']['location']['lat']
    
Lat = pd.DataFrame.from_dict(d, orient='index', columns= ['Lat'])
df_elec_2019['latitude'] = Lat
```

```python
def get_center_latlong(df_elec_2019):
    # get the center of my map for plotting
    centerlat = (df_elec_2019['Lat'].max() + df_elec_2019['Lat'].min()) / 2
    centerlong = (df_elec_2019['Lon'].max() + df_elec_2019['Lon'].min()) / 2
    return centerlat, centerlong
```

```python
center = get_center_latlong(df_elec_2019)
m = folium.Map(location=center, zoom_start=10)
df = df_elec_2019
# go through each home in set, make circle, and add to map.
for i in range(len(df_elec_2019)):
    folium.Circle(
        location=[df_elec_2019.iloc[i]['Lat'], df_elec_2019.iloc[i]['Lon']],
        radius=10,
        fill=True, # Set fill to True
    ).add_to(m)

# Same as before, we save it to file
m.save('circle_map.html')
m
```
